```python
import pandas as pd

'''
Reading Files, Selecting Columns, and Summarizing
'''

# can read a file from local computer or directly from a URL
pd.read_table('data/chipotle.tsv')

chipotleData=pd.read_table('data/chipotle.tsv', index_col="order_id",dtype={'choice_description':list,'item_price':str})

```

```python
chipotleData.columns

chipotleData.dtypes

chipotleData.shape

chipotleData['item_PriceFloat'] = 0

chipotleData.info()
```

```python
floatPriceList=[]
for item in chipotleData.item_price:
    floatPrice=float(item[1:len(item)])
    print "Original: "+item+" new price: "+ str(floatPrice)
    floatPriceList.append(floatPrice)
    
```

```python
chipotleData.item_PriceFloat=floatPriceList

chipotleData.describe()

chipotleData.mean()

chipotleData.item_PriceFloat.value_counts()
#By Frequency, the most popular price paid was $8.75 and $11.25. Looks like we could perhaps build a combo menu to convince guests to order a side of chips with the burrito+drink!

```
